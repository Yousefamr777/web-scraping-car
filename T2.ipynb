# Introduction 
Data withdraw

```python
from bs4 import BeautifulSoup as bs 
from urllib.request import urlopen

f = open("car.csv", "w", encoding="utf-8")
header = "car_title, car_rating, car_price, \n"
f.write(header)

url_base = "https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D=&list_price_max=&maximum_distance=20&zip="
page_num = 0
while page_num <= 60:
    url = url_base + str(page_num)
    client = urlopen(url)
    html = client.read()
    client.close()
    soup = bs(html, "html.parser")
    container = soup.find_all("div", {"class": "vehicle-card"})
    if not container:
        break 
    
    for i in container:
        ctitle = i.findAll("h2", {"class": "title"})
        car_title = ctitle[0].text.strip()
        
        crating = i.findAll("span", {"class": "sds-rating__count"})
        car_rating = crating[0].text.strip()if crating else ""
        
        cprice = i.findAll("span", {"class": "primary-price"})
        car_price = cprice[0].text.strip().replace(",", "")
        
        
        if car_rating:
            f.write(car_title + ", " + car_rating + ", " + car_price +"\n")
            print(car_title + ", " + car_rating + ", " + car_price )
            print()         
    page_num += 1

f.close()
```

# 1- What are the most common cars?

```python
df['car_title'].value_counts()
```

```python
#    
common_cars = df['car_title'].value_counts()

#      (  10 )
top_n = 10
top_cars = common_cars.head(top_n)

#   
plt.figure(figsize=(10, 6))
top_cars.plot(kind='bar', color='skyblue')
plt.title('Most Common Cars')
plt.xlabel('Car Model')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 2-What is the higher assessment car?

```python
df[' car_rating'].value_counts()
```

```python
#    
common_cars = df[' car_rating'].value_counts()

#      (  10 )
top_n = 10
top_cars = common_cars.head(top_n)

#   
plt.figure(figsize=(10, 6))
top_cars.plot(kind='bar', color='skyblue')
plt.xlabel('Car Title')
plt.ylabel('Car Rating')
plt.title('Highest Rated Car')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
